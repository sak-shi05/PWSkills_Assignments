```python
Cost_Price= int(input("Enter Cost Price(Rs): "))
if Cost_Price>100000:
    Road_Tax= Cost_Price*0.15
    print("Tax to be paid is {}".format(Road_Tax))
elif Cost_Price>50000 and Cost_Price<=100000:
    Road_Tax= Cost_Price*0.10
    print("Tax to be paid is {}".format(Road_Tax))
else:
    Road_Tax= Cost_Price*0.05
    print("Tax to be paid is {}".format(Road_Tax))
```



Python while loop is used to run a block code until a certain condition is met.

Example: program to display numbers from 1 to 5


```python
i = 1
n = 5
while i <= n:
    print(i)
    i = i + 1
```

```python
n= int(input('Enter the number of rows: '))
i=1
while i<=n:
    j=1
    while j<=i:
        print(j, end=' ')
        j+=1
    print('')
    i+=1
```

```python
n=int(input('Enter no: '))
i=n
while i>=0:
    j=0
    while j<=i:
        print(j, end=' ')
        j=j+1
    i=i-1
    print("\r")
```

```python
n=int(input('Enter no: '))
k = 2 * n - 2
i=n+1
while i>=0:
    j=1
    while j<=k:
        print(end=" ")
        j+=1
    k = k + 1
    j=1
    while j<=i+1:
        print("*", end=" ")
        j+=1
    i-=1
    print("")
```
