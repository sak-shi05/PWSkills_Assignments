```python
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,22, 61, 34)}, 
         [56, 'data science'], 'Machine Learning']
list2 = []
for i in list1:
    if type(i) == int:
        list2.append(i)
    elif type(i) ==list:
        for j in i:
            if type(j)==int:
                list2.append(j)
    elif type(i) ==set:
        for j in i:
            if type(j)==int:
                list2.append(j)
    elif type(i) ==tuple:
        for j in i:
            if type(j)==int:
                list2.append(j)
    elif type(i) ==dict:
        for j in i.keys():
            if type(j)==int:
                list2.append(j)
        for k in i.values():
            if type(k)==int:
                list2.append(k)
            elif type(k) ==list:
                for j in k:
                    if type(j)==int:
                        list2.append(j)
            elif type(k) ==set:
                for j in k:
                    if type(j)==int:
                        list2.append(j) 
            elif type(k) ==tuple:
                for j in k:
                    if type(j)==int:
                        list2.append(j)          
        
from functools import reduce
reduce(lambda x,y:x*y,list2)
```
