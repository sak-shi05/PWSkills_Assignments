Five plots in Seaborn along with their usage is as following :
1) Box Plot : 
A box plot is a graphical representation of data that shows the distribution of a dataset by indicating the median, quartiles, and outliers. It is used to identify the spread and skewness of the data.
2) Heatmap : 
A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. It is used to identify patterns in the data, such as correlations or clusters.
3) Pair Plot : 
A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. It is used to identify correlations between variables and to explore the relationships between them.
4) Violin Plot :
A violin plot is a graphical representation of data that combines elements of a box plot and a kernel density plot. It is used to visualize the distribution of a dataset and show its density.
5) Joint Plot : 
A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. It is used to explore the relationship between two variables and to identify correlations between them.

These are just a few examples of the types of plots that can be created using Seaborn. Seaborn offers many other visualizations that can be used to explore and analyze different types of data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
fmri= sns.load_dataset("fmri")
sns.lineplot(x="timepoint",y="signal",data=fmri,style='region',hue='event')
plt.title('Signal over Time for Different Events and Regions')
plt.show()
```

```python
import seaborn as sns
titanic= sns.load_dataset("titanic")
sns.boxplot(x="pclass",y="age",data=titanic)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
diamonds= sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x='price',hue='cut',kde=True)
plt.title('Distribution of Price for various diamond cut')
plt.show()
```

```python
import seaborn as sns
iris= sns.load_dataset("iris")
sns.pairplot(iris,hue="species")
```

```python
import seaborn as sns
flights= sns.load_dataset("flights")
```

```python
flights_matrix = flights.pivot(index="year", columns="month", values="passengers")
flights_matrix
```

```python
import matplotlib.pyplot as plt
sns.heatmap(flights_matrix, 
            cmap="YlGnBu", 
            annot=True, 
            fmt="d", 
            linewidths=.5,
            linecolor="black")
plt.title("Passengers per Month")
plt.show()
```
