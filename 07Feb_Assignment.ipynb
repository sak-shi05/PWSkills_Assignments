```python
Password = input("Please enter your password: ")

def Check_PWD(P):
    UF,LF,DF,SF = 0,0,0,0
    if len(P)==10:
        for i in P:
            if ord(i)>=65 and ord(i)<=90:
                UF+=1
            elif ord(i)>=97 and ord(i)<=122:
                LF+=1
            elif ord(i)>=48 and ord(i)<=57:
                DF+=1
            else:
                SF+=1
    if UF>=2 and LF >=2 and DF >=1 and SF >=3:
        print("Valid Password")
    else:
        print("Invalid Password")
Check_PWD(Password)
```
