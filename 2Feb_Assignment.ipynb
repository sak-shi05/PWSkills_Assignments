```python
For Loop - We use for loop when no of iterations are not known.

        - When condition for breaking the loop is not specified.

# Printing 1 to 5
for i in range(1, 6):  # Here no of iterations are already specified
    print(i)
```

```python
While Loop - We use while loop when the no of iterations are known.
           - When loop has to be breaked if a condition is fulfilled.

# Printing 5 to 1
i = 5
while (i > 0):  # Here i greater than 0 is a condition
    print(i)
    i -= 1
```

```python
# using for loop
Sum=0
Product=1
for i in range(1,11):
    Sum=Sum+i
    Product= Product*i
print("Sum of first 10 natural no. is ",Sum)
print("Product of first 10 natural no. is ",Product)
```

```python
# using for loop
Sum=0
Product=1
i=1
while i<11:
    Sum=Sum+i
    Product= Product*i
    i+=1
print("Sum of first 10 natural no. is ",Sum)
print("Product of first 10 natural no. is ",Product)
```

```python
U=int(input("Enter No of Units used: "))
Total_Unit=U
Total_Amount=0
if U<=100:
    Total_Amount=Total_Amount+ U*4.5    
elif U>100 and U<=200:
    U-=100 
    Total_Amount=450+ U*6    
elif U>200 and U<=300:
    U-=200
    Total_Amount=450+600+ U*10 
elif U>300:
    U-=300
    Total_Amount=450+600+1000+U*20
print("Total amount to be paid for {} units: RS {}".format(Total_Unit,Total_Amount))
```

```python
l=list(range(1,101))
l1=[]
#using for loop
for i in l:
    if (i**3)%4==0 or (i**3)%5==0:
        l1.append(i)
print(l1)
```

```python
l2=[]
i=1
#using while loop
while i<101:
    if (i**3)%4==0 or (i**3)%5==0:
        l2.append(i)
    i+=1
print(l2)
```

```python
string= "I want to become a data scientist"
count=0
for i in string:
    if i.lower() in ['a','e','i','o','u']:
        count+=1
print(f"No of vowels in the given string is : {count}")
```
